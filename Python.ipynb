## Replace f/t values of instant bookable column to No/Yes respectively (hint: inplace = True)

```python
#inplace=True
df['instant_bookable'] = df['instant_bookable'].replace("f", "No")\
                                               .replace("t", "Yes")
```

```python
df['instant_bookable']
```

## Sort dataset by price in descending order

```python
#inplace=True
df.sort_values(['price'], ascending=False)

```

## Fill empty (nan) values of host_response_time to N/A

```python
#inplace=True
df['host_response_time'] = df['host_response_time'].replace(np.nan, "N/A")
```

```python
df
```

## Find how many listings per host_response_time in a pie chart

```python
df.groupby(['host_response_time'], as_index=False)['host_total_listings_count'].sum()
 

 
```

```python
labels = ['N/A', 'A few days','Within A day', 'A few hours','An hour']
sizes = [685, 85, 292, 823, 1796]
explode = (0, 0, 0, 0, 0.2)  # only "explode" the 1st slice (The biggest)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

## Find the average price per city and visualize it in a sorted, bar chart

```python
df.groupby(['city'], as_index=False)['price'].mean() 
```

```python
x = ["Bangkok", "Cape Town", "Hong Kong", "Istanbul", "Mexico City", "New York", "Paris", "Rio de Janeiro", "Rome", "Sydney"]
y = [1673.82, 2101.61, 979.00, 557.06, 2392.82, 151.35, 107.66, 431.37, 87.56,1110.43]
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar( x, y, color ='blue', width = 0.4)
 
#Add label to x axis 
plt.xlabel("List of Cities")

#Add label to y axis
plt.ylabel("Average Price")

#Add title to plot
plt.title("Average Price Per City")


plt.show()
```
